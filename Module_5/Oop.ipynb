### Class and Objects



```python
# base class
class phone:
    category = "Electronocs"
    hello = "Hello"

    # constructor
    def __init__(self,model,battery,camera,battery_percentage=100):
        self.model = model
        self.battery= battery
        self.camera= camera
        self.battery_percentage = battery_percentage

    def charge(self,hour):
        print(f"Charge completed by {hour}")
    def capture(self,photo):
        if(self.battery_percentage)<=0:
            print("No Charge")
        else:
            self.battery_percentage -=photo
            print(f"photo captured in {self.model}")

class Cooling_Mechanism:
    def __init__(self,cooling_method):
        self.cooling_method = cooling_method
    def cooling_on(self):
        print(f"The system is being cool by {self.cooling_method}")
```

```python
## creating object

apple = phone()
print(type(apple))
```

```python
apple = phone()
apple.category = "Super Electronocs"
blueberry = phone()
motorola = phone()

print(apple.category)
print(apple.hello)
print(motorola.category)
print(blueberry.category)
```

```python
apple = phone("Iphone17",3000,40)
blueberry = phone("B-17",4000,30)
motorola = phone("M-17",4500,35)
```

```python
print(apple.camera)
print(apple.category)
print(motorola.camera)
print(motorola.battery_percentage)
apple.capture(10)
print(apple.battery_percentage)
```

## Inheritance

```python
# child/derived class
class SmartPhone(phone):
    def __init__(self,model,battery,camera,processor):
        super().__init__(model,battery,camera)
        self.processor = processor
    def charge(self,hour):
        print("fast charging in process")
        super().charge(hour)
```

```python
pro = SmartPhone("X",5000,100,"SnapDragon")

print(pro.model)
pro.charge(1)
```

### Multiple inheritance

```python
class SmartPhone_CoolingMode(SmartPhone,Cooling_Mechanism):
    def __init__(self,model,battery,camera,processor,cooling_method):
        SmartPhone.__init__(self,model,battery,camera,processor)
        Cooling_Mechanism.__init__(self,cooling_method)
```

```python
pro_cooling = SmartPhone_CoolingMode("Y",5000,100,"SD","Nitrogen")

print(pro_cooling.processor)
print(pro_cooling.battery)
print(pro_cooling.cooling_method)
print(pro_cooling.cooling_on())
print(pro_cooling.charge(1))
```

### Polymorphsim

```python
class Camera:
    def __init__(self,name):
        self.name = name
    def capture(self):
        print("a photo is captured")
        
```

```python
class Smart(Camera):
    def __init__(self,name,resolution):
        super().__init__(name)
        self.resolution=resolution
    # method overriding
    def capture(self):
        print("Photo is captured by a Phone")


class DSLR(Camera):
    def __init__(self,name,resolution):
        super().__init__(name)
        self.resolution=resolution

    def capture(self):
        print("Photo is captured by a DSLR")

class Drone(Camera):
    def __init__(self,name,resolution):
        super().__init__(name)
        self.resolution=resolution

    def capture(self):
        print("Photo is captured by a Drone")

phone = Smart("Phone",30)
dslr = DSLR("DSLA",40)
drone = Drone("Drone",35)

phone.capture()
dslr.capture()
drone.capture()

```

## Encapsulation

```python
class Mobile:
    def __init__(self,name,model,imei):
        self.__name=name
        self.__model=model
        self.__imei= imei # private
    def charge(self):
        Print("Phone is charging")

    def imei_getter(self):
        return self.__imei
    def name_getter(self):
        return self.__name
    def model_getter(self):
        return self.__model
    def name_setter(self,name):
        self.__name = name 

iphone = Mobile("Iphone","17","1xkaf1")
print(iphone.name_getter())
print(iphone.model_getter())
print(iphone.imei_getter())

iphone.name_setter("Dip")
print(iphone.name_getter())
```

## Abstraction

```python
from abc import ABC, abstractmethod

class Telephone(ABC):

    @abstractmethod
    def make_call(self):
        pass

    @abstractmethod
    def capture_photo(self):
        pass


class Sphone(Telephone):
    def make_call(self):
        print("Making a call using Sphone")

    def capture_photo(self):
        print("Capturing photo using Sphone")


class Iphone(Telephone):
    def make_call(self):
        print("Making a call using Iphone")

    def capture_photo(self):
        print("Capturing photo using Iphone")


# Now it will work
ip = Iphone()
ip.make_call()
ip.capture_photo()
```
