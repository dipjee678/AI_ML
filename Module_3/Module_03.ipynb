### Seek

```python
with open("./sample_data/sample.txt","r") as file:
  print(file.tell())
  file.seek(5)
  print(file.tell())
  file.seek(0)
  print(file.tell())
```

## Practice 1

```python
from functools import reduce
with open("./sample_data/sample.txt","r") as file:
  string_list = file.readlines()
  total_line = len(string_list)

  print(string_list)
  print(total_line)

  number_of_word = list(map(lambda x : len(x.split()), string_list))
  print(number_of_word)

  total_numbers_of_words = reduce(lambda x,y : x+y, number_of_word)
  print(total_numbers_of_words)

  # cleaning process

  string = list(map(str.strip,string_list))
  print(string)

  # space Delete
  string_list = list(map(lambda x : x.replace(" ",""),string))
  print(string_list)

  # Number of Character
  number_of_Character = list(map(lambda x : len(x), string_list))
  print(number_of_Character)

  # Total Number of Character
  total_numbers_of_Character = reduce(lambda x,y : x+y, number_of_Character)
  print(total_numbers_of_Character)

with open("./sample_data/counter_of_string.txt","w") as file:
  file.write(f"Total line :{total_line}\n number_of_words {number_of_word}\n total_numbers_of_words {total_numbers_of_words}\n number_of_Character {number_of_Character}\n total_numbers_of_Character {total_numbers_of_Character} ")


```
