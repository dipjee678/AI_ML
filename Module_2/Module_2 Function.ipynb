## Different type of argument


```python
def adder(a,b):
    print(f"{a},{b}")

adder(10,20)
```

```python
def adder(*args):
    print(args)


adder(10,30,40,20)
```

```python
def adder(*args):
    for i in args:
        print(i)


adder(10,30,40,20)
```

### Keyword Argument


```python
def adder(a,b,c):
    print(f"{a},{b},{c}")

adder(a=10,c=30,b=20)
```

```python
def adder(**kwargs):
        print(kwargs)


adder(a=10,c=30,b=20)
```

```python
def adder(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
      


adder(a=10,c=30,b=20)
```

## Genaratror

```python
lst = [x for x in range(500)]

def data_loader(chunk_size,lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]

x= data_loader(5,lst)

print(type(x))
print(x)


print(next(x))
print(next(x))
```

## Lambda Function

```python
import pandas as pd

df = pd.DataFrame({'temperature (F)':[32,212,98.6,68]})
df
df['temperature (c)']=df['temperature (F)'].apply(lambda x : (x-32)*5/9)
df
```

```python
import pandas as pd

df = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})

df

df['A'] = df['A'].map(lambda x:x ** 0.5)
df
```
